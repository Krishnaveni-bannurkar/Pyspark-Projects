```python
import pyspark
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, avg

spark = SparkSession.builder.appName("Experience level and salary analysis").getOrCreate()

read_data = spark.read.csv("/Users/Krishnaveni/Downloads/ds_salaries.csv",  header=True, inferSchema=True)

#get the total and avg values of the employees based on experience level and job title
aggrigated_avg_sal = read_data.groupBy("experience_level", "job_title") \
                        .agg(avg("salary_in_usd").alias("Avg_Salary_USD") , sum("salary_in_usd").alias("Total_Salary_USD"))\
                        .orderBy("Avg_Salary_USD", ascending=False)
print("aggrigated_avg_sal based on the experience_leveland job_title  ")
aggrigated_avg_sal.show(5)


#Compare the salaries of different employment types

Compare_salaries_employment_types = read_data.groupBy("employment_type")\
                                    .agg(avg("salary_in_usd").alias("Avg_Salary_USD"),sum("salary_in_usd").alias("Total_Salary_USD"))\
                                    .orderBy("Avg_Salary_USD", ascending=False)
print("Compare_salaries based on the employment_types")
Compare_salaries_employment_types.show(5)


#impact of remote work on salaries
impact_of_remoteWork_on_Sal = read_data.groupBy("remote_ratio")\
                                    .agg(avg("salary_in_usd").alias("Avg_Salary_USD"),sum("salary_in_usd").alias("Total_Salary_USD"))\
                                    .orderBy("Avg_Salary_USD", ascending=False)
print("impact_of_remoteWork on Salaries")
impact_of_remoteWork_on_Sal.show(5)


#Compare salaries across different countries
Compare_salaries_Diferent_countries = read_data.groupBy("company_location")\
                                    .agg(avg("salary_in_usd").alias("Avg_Salary_USD"),sum("salary_in_usd").alias("Total_Salary_USD"))\
                                    .orderBy("Avg_Salary_USD", ascending=False)
print("Compare_salaries based on the Diferent_countries")
Compare_salaries_Diferent_countries.show(5)


Compare_salaries_Diferent_countries_Size = read_data.groupBy("company_size")\
                                    .agg(avg("salary_in_usd").alias("Avg_Salary_USD"),sum("salary_in_usd").alias("Total_Salary_USD"))\
                                    .orderBy("Avg_Salary_USD", ascending=False)
print("Compare_salaries based on the Diferent_countries_size")
Compare_salaries_Diferent_countries_Size.show(5)


#aggrigated_avg_sal.write.csv("/Users/Krishnaveni/results/aggrigated_avg_sal.csv", header=True)
#Compare_salaries_employment_types.write.csv("/Users/Krishnaveni/results/Compare_salaries_employment_types.csv", header=True)
#impact_of_remoteWork_on_Sal.write.csv("/Users/Krishnaveni/results/impact_of_remoteWork_on_Sal.csv", header=True)
#Compare_salaries_Diferent_countries.write.csv("/Users/Krishnaveni/results/Compare_salaries_Diferent_countries.csv", header=True)
#Compare_salaries_Diferent_countries_Size.write.csv("/Users/Krishnaveni/results/Compare_salaries_Diferent_countries_Size.csv", header=True)



```
